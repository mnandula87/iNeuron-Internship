# Find Sales Trend Month Wise, Year wise, yearly_monthwise

```python
import datetime
```

```python
df['year'] = pd.DatetimeIndex(df['Invoice Date']).year
```

```python
df['month'] = pd.DatetimeIndex(df['Invoice Date']).month
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
##Grouping the year to get the sales across the category
df_year = df.groupby(['year'])['Sales Amount'].sum().reset_index()
df_year
```

```python
df_year_month = df.groupby(['year','month'])['Sales Amount'].sum().reset_index()
df_year_month
```

```python
df_month = df.groupby(['month'])['Sales Amount'].sum().reset_index()
df_month
```

# Data Visualization

# Yearly Sales Data

```python
sns.set(style="darkgrid")
plt.figure(figsize=(8, 8))
sns.barplot(x="year", y="Sales Amount", data=df_year)
```

# Yearly Month Wise Sales Data

```python
sns.set(style="darkgrid")
plt.figure(figsize=(8, 8))
sns.barplot(x="month", y="Sales Amount", hue="year", data=df_year_month)
```

Here the reason why 2018 sales are low when compared to the other years is evident because 2018 sales after the month March are missing.
Hence we can attribute low sales to the missing data

# Monthly Sales Data

```python
sns.set(style="darkgrid")
plt.figure(figsize=(8, 8))
sns.barplot(x="month", y="Sales Amount", data=df_month)
```

# Top 10 sales rep as per sales

```python
##Top 10 sales rep as per sales
df.groupby(['Sales Rep'])['Sales Amount'].sum().sort_values(ascending=False).iloc[:10].plot.barh()
plt.xlabel("Sales")
plt.ylabel("Sales Rep ID")
plt.title("Top 10 Sales rep as per sales for three years")
plt.show()
```

# Least 10 sales rep as per sales

```python
#Least 10 sales rep as per sales
df.groupby(['Sales Rep'])['Sales Amount'].sum().sort_values(ascending=False).iloc[-10:].plot.barh()
plt.xlabel("Sales")
plt.ylabel("Sales Rep ID")
plt.title("Least 10 Sales rep as per sales for three years")
plt.show()
```

The best performer in sales award goes to sales rep with ID 141 having 25 million sales under his belt.
